```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARMA
from IPython.display import Image
from datetime import datetime, timedelta
register_matplotlib_converters()

import warnings
warnings.filterwarnings("ignore")
```

# Ice Cream Production Data

```python
#read data
df_ice_cream = pd.read_csv('../data/ice_cream.csv')
```

```python
df_ice_cream.head()
```

```python
#rename columns to something more understandable
df_ice_cream.rename(columns={'DATE':'date', 'IPN31152N':'production'}, inplace=True)
```

```python
#convert date column to datetime type
df_ice_cream['date'] = pd.to_datetime(df_ice_cream.date)
```

```python
#set date as index
df_ice_cream.set_index('date', inplace=True)
```

```python
#just get data from 2010 onwards
start_date = pd.to_datetime('2010-01-01')
df_ice_cream = df_ice_cream[start_date:]
```

```python
#show result
df_ice_cream.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df_ice_cream.production)
plt.title('Ice Cream Production over Time', fontsize=20)
plt.ylabel('Production', fontsize=16)
for year in range(2011,2021):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
plt.axhline(df_ice_cream.mean()[0], color='k', linestyle='--', alpha=0.2)    
```

# ACF

```python
acf_plot = plot_acf(df_ice_cream.production, lags=100)
```

## Based on decaying ACF, we are likely dealing with an Auto Regressive process

# PACF

```python
pacf_plot = plot_pacf(df_ice_cream.production)
```

## Based on PACF, we should start with an Auto Regressive model with lags 1, 2, 3, 10, 13

# Get training and testing sets

```python
train_end = datetime.datetime(2016,12,30)
test_end = datetime.datetime(2020,1,1)

train_data = df_ice_cream[:train_end].squeeze()
test_data = df_ice_cream[train_end:test_end].squeeze()
```

# Fitting the AR Model
For simplicity we are taking the order as 3

```python
#create the model
model = ARMA(train_data, order=(4,0))
```

```python
import time as t
#fit the model
start = t.time()
model_fit = model.fit()
end = t.time()
print('Model Fitting Time:', end - start)
```

```python
#summary of the model
print(model_fit.summary())
```

## So the ARMA(1,0) model is:
### $\hat{y_t} = 102.5717 + 0.9701y_{t-1}$

```python
#get prediction start and end dates
pred_start_date = test_data.index[0]
pred_end_date = test_data.index[-1]
```

```python
#get the predictions and residuals
predictions = model_fit.predict(start=pred_start_date, end=pred_end_date)
```

```python
residuals = test_data - predictions
```

```python
plt.figure(figsize=(10,4))
plt.plot(residuals)
plt.title('Residuals from AR Model', fontsize=20)
plt.ylabel('Error', fontsize=16)
plt.axhline(0, color='r', linestyle='--', alpha=0.2)
```

## Resisdual clearly show a pattern in the erros which is not desirable and this means that we were not able to capture the full meaning in the data

```python
plt.figure(figsize=(10,4))

plt.plot(test_data)
plt.plot(predictions)

plt.legend(('Data', 'Predictions'), fontsize=16)
```

```python
print('Root Mean Squared Error:', np.sqrt(np.mean(residuals**2)))
```

```python
print('Mean Absolute Percent Error:', round(np.mean(abs(residuals/test_data)),4))
```

## Rolling Forecast Origin

## Idea: Predict one month in advance each time

### Train on months 1, 2, 3, ..., K-2 -> predict month k-1
### Train on months 1, 2, 3, ..., K-1 -> predict month k

.........

### Averaging all prediction to get the overall performance of the model

```python
prediction_rolling = pd.Series()
for end_date in test_data.index:
    train_data = df_ice_cream[:end_date - timedelta(days=1)]
    model = ARMA(train_data, order=(3,0))
    model_fit = model.fit()
    pred = model_fit.predict(end_date)
    prediction_rolling.loc[end_date] = pred.loc[end_date]
```

```python
residuals = test_data - prediction_rolling
```

```python
plt.figure(figsize=(10,4))
plt.plot(residuals)
plt.title('Residuals from AR Model', fontsize=20)
plt.ylabel('Error', fontsize=16)
plt.axhline(0, color='r', linestyle='--', alpha=0.2)
```

## Although there is some sort of a volatality in the residuals but there's no as such systemic pattern as we saw before

```python
plt.figure(figsize=(10,4))

plt.plot(test_data)
plt.plot(prediction_rolling)

plt.legend(('Data', 'Predictions'), fontsize=16)
```

```python
print('Root Mean Squared Error:', np.sqrt(np.mean(residuals**2)))
```

```python
print('Mean Absolute Percent Error:', round(np.mean(abs(residuals/test_data)),4))
```

## As we can observe from the previous results the performance has improve drastically when we are using a rolling forecast Origin!!!
